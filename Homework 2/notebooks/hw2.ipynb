## Contans

```python
# [k_B] = eV/K
k_B = 8.617333262145e-05
# [T_c] = K
T_c = 0.001
# [m] = eV
m = 931e03
# [hbar] = eV * s
hbar = 6.582119569e-16
```

## Plot fugacity

```python
def fugacity(n, m):
    return k_B * T_c * 2 * m / hbar**2 * n**(-2/3)
```

```python
fig, axes = plt.subplots(figsize=(12,9))

n = np.linspace(0,10,101)*6e23

axislabelsize = 19
axisticksize = 19
offsetsize = 16

axes.plot(n, fugacity(n, m=938e06),
          color='tab:red', lw=4)
axes.plot(n, fugacity(n, m=511e03),
          color='tab:green', lw=4)
axes.plot(n, fugacity(n, m=1931e06),
          color='tab:blue', lw=4)

axes.set_xlabel('$N/V$', fontsize=axislabelsize)
axes.set_ylabel('z', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)
axes.xaxis.get_offset_text().set_fontsize(offsetsize)
axes.yaxis.get_offset_text().set_fontsize(offsetsize)


plt.show()
```
